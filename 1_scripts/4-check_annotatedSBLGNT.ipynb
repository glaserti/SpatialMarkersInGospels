```python
def df_constructor(cltk_object):
    ll = []
    for chap in cltk_object:
        l = []
        for e in cltk_object[chap]:
            #total_words = len(cltk_object[chap].lemmata) # calculating the number of words per chapter
            b,c = chap.split(" ")
            l = [b, c, e.index_sentence, e.index_token, e.lemma.lower(), e.string]
            ll.append(l)

                    # separating chap ("Mark 1") into book and chapter-nr ("Mark", "1")
                    #b,c = chap.split(" ")

                    #extracting chapter, sentence & token ix, lemma
                    #l = [b, c, total_words, e.index_sentence, e.index_token, e.lemma, kwic]
                    #ll.append(l)

    #creating a dataframe 
    df = pd.DataFrame(ll, columns=['book', 'chap', 'sentence_Ix', 'token_Ix', 'lemma', "word"])
    return(df)
```

```python
def saveAsCSV(bookname, cltk_object):
    filename = "data_annotated/SBLGNT/" + bookname + "_word_lemma.tsv"
    book_df = df_constructor(cltk_object)
    book_df.to_csv(filename, index=False, sep="\t")
    return(book_df)
```
